## 9Ô∏è‚É£ Manejo de Valores Nulos

```python
# Verificar valores nulos
df.isnull().sum()

```

```python
df.head()
```

```python
# Rellenar valores nulos con la media de la columna
df['Edad'].fillna(df['Edad'].mean().round(0), inplace=True)
df
```

## üîü Aplicaci√≥n de Funciones Personalizadas

```python
# Crear una funci√≥n para clasificar empleados seg√∫n experiencia
def clasificar_experiencia(anios):
    if anios < 3:
        return 'Junior'
    elif anios < 10:
        return 'Mid-Level'
    else:
        return 'Senior'

df['Nivel_Experiencia'] = df['A√±os_Experiencia'].apply(clasificar_experiencia)
df
```
